```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline

from sklearn.linear_model import LinearRegression
```

```python
# importing the dataset csv file
dataframe = pd.read_csv("USA-IPC.csv")
# viewing the form of the dataframe
print(dataframe.info())
# viewing the first 5 rows, or the head, of the dataset
print(dataframe.head())
# viewing the last 5 rows, or the tail, of the dataset
print(dataframe.tail())
```

```python
# viewing the regression chart of the dataframe
sb.regplot(x = "Year", y = "Income", data = dataframe)
# print the correlation coefficient (r-value)
print(dataframe["Year"].corr(dataframe["Income"]))
# --> correlation coefficient of 0.9933898070663643 indicates a strong positive correlation between Year vs Income
```

```python
linearRegressionObject = LinearRegression()
linearRegressionObject.fit(dataframe[["Year"]].values, dataframe.Income)
```

```python
linearRegressionObject.predict([[2023]])
```
